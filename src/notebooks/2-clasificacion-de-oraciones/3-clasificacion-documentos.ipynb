```python
from nltk import word_tokenize, FreqDist
```

```python
df = pd.read_csv('/tmp/datasets/email/csv/spam-apache.csv', names=['clase', 'contenido'])
df['tokens'] = df['contenido'].apply(word_tokenize)
df
```

```python
all_words = FreqDist([w for tokenlist in df['tokens'].values for w in tokenlist])
top_words = all_words.most_common(200)

def document_atributos(document):
    document_words = set(document)
    atrib = {}
    for word in top_words:
        atrib[f'contains({word})'] = (word in document_words)
    return atrib
```

```python
document_atributos(df.tokens.values[0])
```

```python
import random
```

```python
fset = [(document_atributos(texto), clase) for texto, clase in zip(df['tokens'].values, df['clase'].values)]
random.shuffle(fset)
```

```python
train, test = fset[200:], fset[:200]
```

```python
import nltk
```

```python
classifier = nltk.NaiveBayesClassifier.train(train)
```

```python
nltk.classify.accuracy(classifier, test)
```
